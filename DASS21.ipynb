```python
def dass21q(questions,dq,aq,sq,sdais,ansrange,dass21choices):
  dscore =0
  ascore =0
  sscore =0

  dscoreis =sdais[3]
  ascoreis =sdais[4]
  sscoreis =sdais[5]

  for i in questions:
    if i in dq:
     a = pyip.inputNum(str(i)+"\n"+str(dass21choices)+"\n\n",min=0,max=3)
     dscore= dscore+a
    elif i in aq:
     a = pyip.inputNum(str(i)+"\n"+str(dass21choices)+"\n\n",min=0,max=3)
     ascore= ascore+a
    elif i in sq:
     a = pyip.inputNum(str(i)+"\n"+str(dass21choices)+"\n\n",min=0,max=3)
     sscore= sscore+a
    else:
      print("Bad luck")


  print (dscoreis, dscore)
  print (ascoreis, ascore)
  print (sscoreis,sscore)
  print ("\n")
  dass21cat(dscore,ascore,sscore,ansrange,sdais)
```

```python
def dass21cat(dscore,ascore,sscore,ansrange,sdais):
  dans=sdais[0]
  aans=sdais[1]
  sans=sdais[2]

  nl =ansrange[0]
  mil =ansrange[1]
  mod =ansrange[2]
  svr =ansrange[3]
  exsvr =ansrange[4]

  if 0<=sscore<=10:
    print(sans,nl)
  elif 11<=sscore<=18:
    print(sans,mil)
  elif 19<=sscore<=26:
    print(sans,mod)
  elif 27<=sscore<=34:
    print(sans,svr)
  elif 35<=sscore<=42:
    print(sans,exsvr)

  if 0<=dscore<=9:
    print(dans,nl)
  elif 10<=dscore<=12:
    print(dans,mil)
  elif 13<=dscore<=20:
    print(dans,mod)
  elif 21<=dscore<=27:
    print(dans,svr)
  elif 28<=dscore<=42:
    print(dans,exsvr)

  if 0<=ascore<=6:
    print(aans,nl)
  elif 7<=ascore<=9:
    print(aans,mil)
  elif 10<=ascore<=14:
    print(aans,mod)
  elif 15<=ascore<=19:
    print(aans,svr)
  elif 20<=ascore<=42:
    print(aans,exsvr)
```
