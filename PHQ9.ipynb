```python
def phq9():
  print("လွန်ခဲ့သော ရက်သတ္တ (၂) ပါတ်အတွင်း အောက်ဖော်ပြပါပြဿနာများကို \
  မည်သည့်အတိုင်းအတာအထိ ခံစားခဲ့ရပါသနည်း\n\nအဖြေ 0 မှ 3 ကြားရွေး ရိုက်ထည့်ပါ၊")


  a1 = pyip.inputNum(str(q1) +"\n"+ str(phq9choices)+"\n\n",min=0,max=3)
  a2 = pyip.inputNum(str(q2) +"\n"+ str(phq9choices)+"\n\n",min=0,max=3)
  a3 = pyip.inputNum(str(q3) +"\n"+ str(phq9choices)+"\n\n",min=1,max=3)
  a4 = pyip.inputNum(str(q4) +"\n"+ str(phq9choices)+"\n\n",min=1,max=3)
  a5 = pyip.inputNum(str(q5) +"\n"+ str(phq9choices)+"\n\n",min=1,max=3)
  a6 = pyip.inputNum(str(q6) +"\n"+ str(phq9choices)+"\n\n",min=1,max=3)
  a7 = pyip.inputNum(str(q7) +"\n"+ str(phq9choices)+"\n\n",min=1,max=3)
  a8 = pyip.inputNum(str(q8) +"\n"+ str(phq9choices)+"\n\n",min=1,max=3)
  a9 = pyip.inputNum(str(q9) +"\n"+ str(phq9choices)+"\n\n",min=1,max=3)

  answers = [a1,a2,a3,a4,a5,a6,a7,a8,a9]

  if all(i <= 4 for i in answers):
    totalscore = int(a1+a2+a3+a4+a5+a6+a7+a8+a9)
    print ("Total score is : {}" .format(totalscore))
    phq9category(totalscore)
  else:
    print("Answer should be between 0 to 3. Result can't be produced, Plz try again.")
```

တတ်ကျွမ်းနားလည်သူများနှင့် ဆွေးနွေးကုသပါ။

```python
def phq9category(totalscore):
 if 1<=totalscore<=4:
   print("သင့်တွင် စိတ်ကျဝေဒနာလက္ခဏာများ မတွေ့ရှိပါ။ ")
 elif 5<=totalscore <=14: 
   print("သင့်တွင် စိတ်ကျဝေဒနာဖြစ်နိုင်ချေ ရှိနေပါသည်။ ဆွေးနွေးရန် လိုအပ်ပါသည်။")
 elif 15<=totalscore<=27:
   print("သင့်တွင် စိတ်ကျဝေဒနာဖြစ်နိုင်ချေ များနေပါသည်။\nတတ်ကျွမ်းနားလည်သူများနှင့် ဆွေးနွေးကုသပါ။")
 else:
  print("invalid answer")
```
